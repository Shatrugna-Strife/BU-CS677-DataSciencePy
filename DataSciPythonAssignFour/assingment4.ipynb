```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("heart_failure_clinical_records_dataset.csv" )
data = data[["creatinine_phosphokinase", "serum_creatinine", "serum_sodium", "platelets", "DEATH_EVENT"]]
data
```

```python
df_0 = data[data["DEATH_EVENT"] == 0]
df_0
```

```python
df_1 = data[data["DEATH_EVENT"] == 1]
df_1
```

```python
def group_execute(G3_0,death_event):
    sse = []
    X_train_0, X_test_0,y_train_0, y_test_0 = train_test_split(G3_0.iloc[:,0],G3_0.iloc[:,1] ,random_state=124, test_size=0.5, shuffle=True)
    lin_reg = LinearRegression()
    lin_reg.fit(X_train_0.to_numpy().reshape(-1, 1),y_train_0.to_numpy().reshape(-1, 1))
    print("DEATH_EVENT "+death_event+" Linear Regression y = ax + b")
    print("a=",lin_reg.coef_[0][0], "b=",lin_reg.intercept_[0])

    y_pred = lin_reg.predict(X_test_0.to_numpy().reshape(-1, 1))
    print("SSE =", np.sum((y_pred-y_test_0.to_numpy())**2))

    sse.append(np.sum((y_pred-y_test_0.to_numpy())**2))

    x_arg = X_test_0.to_numpy().argsort()

    plt.plot(X_test_0, y_test_0.to_numpy().reshape(-1, 1), "b.")
    plt.plot(X_test_0.to_numpy()[x_arg[::]],y_pred[x_arg[::]],"k--")
    plt.xlabel(G3_0.columns[0])
    plt.ylabel(G3_0.columns[1])
    plt.title("DEATH_EVENT "+death_event+" Linear Regression y = ax + b")
    plt.show()


    poly = PolynomialFeatures(2, include_bias=False)
    X_new = poly.fit_transform(X_train_0.to_numpy().reshape(-1, 1))
    X_new
    lin_reg = LinearRegression()
    lin_reg.fit(X_new,y_train_0.to_numpy().reshape(-1, 1))
    print("DEATH_EVENT "+death_event+" quadratic y = ax2 + bx + c")
    print("a=",lin_reg.coef_[0][0],"b=",lin_reg.coef_[0][1], "c=",lin_reg.intercept_[0])

    X_plot_new = poly.fit_transform(X_test_0.to_numpy().reshape(-1, 1))
    y_pred = lin_reg.predict(X_plot_new)
    print("SSE =", np.sum((y_pred-y_test_0.to_numpy())**2))

    sse.append(np.sum((y_pred-y_test_0.to_numpy())**2))

    x_arg = X_test_0.to_numpy().argsort()

    plt.plot(X_test_0, y_test_0.to_numpy().reshape(-1, 1), "b.")
    plt.plot(X_test_0.to_numpy()[x_arg[::]],y_pred[x_arg[::]],"k--")
    plt.xlabel(G3_0.columns[0])
    plt.ylabel(G3_0.columns[1])
    plt.title("DEATH_EVENT "+death_event+" quadratic y = ax2 + bx + c")
    plt.show()



    poly = PolynomialFeatures(3, include_bias=False)
    X_new = poly.fit_transform(X_train_0.to_numpy().reshape(-1, 1))
    X_new
    lin_reg = LinearRegression()
    lin_reg.fit(X_new,y_train_0.to_numpy().reshape(-1, 1))
    print("DEATH_EVENT "+death_event+" cubic spline  y = ax3 + bx2 + cx + d")
    print("a=",lin_reg.coef_[0][0],"b=",lin_reg.coef_[0][1],"c=",lin_reg.coef_[0][2], "d=",lin_reg.intercept_[0])

    X_plot_new = poly.fit_transform(X_test_0.to_numpy().reshape(-1, 1))
    y_pred = lin_reg.predict(X_plot_new)
    print("SSE =", np.sum((y_pred-y_test_0.to_numpy())**2))

    sse.append(np.sum((y_pred-y_test_0.to_numpy())**2))

    x_arg = X_test_0.to_numpy().argsort()

    plt.plot(X_test_0, y_test_0.to_numpy().reshape(-1, 1), "b.")
    plt.plot(X_test_0.to_numpy()[x_arg[::]],y_pred[x_arg[::]],"k--")
    plt.xlabel(G3_0.columns[0])
    plt.ylabel(G3_0.columns[1])
    plt.title("DEATH_EVENT "+death_event+" cubic spline  y = ax3 + bx2 + cx + d")
    plt.show()


    lin_reg = LinearRegression()
    lin_reg.fit(np.log(X_train_0.to_numpy().reshape(-1, 1)),y_train_0.to_numpy().reshape(-1, 1))
    print("DEATH_EVENT "+death_event+" GLM - generalized linear model  y = a log x + b")
    print("a=",lin_reg.coef_[0][0], "b=",lin_reg.intercept_[0])

    y_pred = lin_reg.predict(np.log(X_test_0.to_numpy().reshape(-1, 1)))

    sse.append(np.sum((y_pred-y_test_0.to_numpy())**2))
    print("SSE =", np.sum((y_pred-y_test_0.to_numpy())**2))

    x_arg = X_test_0.to_numpy().argsort()

    plt.plot(X_test_0, y_test_0.to_numpy().reshape(-1, 1), "b.")
    plt.plot(X_test_0.to_numpy()[x_arg[::]],y_pred[x_arg[::]],"k--")
    plt.xlabel(G3_0.columns[0])
    plt.ylabel(G3_0.columns[1])
    plt.title("DEATH_EVENT "+death_event+" GLM - generalized linear model  y = a log x + b")
    plt.show()


    lin_reg = LinearRegression()
    lin_reg.fit(np.log(X_train_0.to_numpy().reshape(-1, 1)),np.log(y_train_0.to_numpy().reshape(-1, 1)))
    print("DEATH_EVENT "+death_event+" GLM - generalized linear model  log y = a log x + b")
    print("a=",lin_reg.coef_[0][0], "b=",lin_reg.intercept_[0])

    y_pred = np.exp(lin_reg.predict(np.log(X_test_0.to_numpy().reshape(-1, 1))))
    print("SSE =", np.sum((y_pred-y_test_0.to_numpy())**2))

    sse.append(np.sum((y_pred-y_test_0.to_numpy())**2))

    x_arg = X_test_0.to_numpy().argsort()

    plt.plot(X_test_0, y_test_0.to_numpy().reshape(-1, 1), "b.")
    plt.plot(X_test_0.to_numpy()[x_arg[::]],y_pred[x_arg[::]],"k--")
    plt.xlabel(G3_0.columns[0])
    plt.ylabel(G3_0.columns[1])
    plt.title("DEATH_EVENT "+death_event+" GLM - generalized linear model  log y = a log x + b")
    plt.show()
    return sse
```

```python
i = 1
for temp in [["creatinine_phosphokinase", "platelets"], ["platelets", "serum_sodium"], ["serum_sodium", "serum_creatinine"], ["platelets", "serum_creatinine"]]:
    print("Group -",i,temp)
    sse_0 = group_execute(df_0[temp], "0")
    sse_1 = group_execute(df_1[temp], "1")
    data = {'Model': ['y = ax + b', 'y = ax2 + bx + c', 'y = ax3 + bx2 + cx + d', 'y = a log x + b', 'log y = a log x + b'],
        'SSE (death event=0)': sse_0,
        'SSE (death event=1)': sse_1}
    print(pd.DataFrame(data, index=[""]*5))
    print("")
    print("")
    i+=1
```
