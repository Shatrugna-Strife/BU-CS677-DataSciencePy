```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
%matplotlib inline
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
df_sheet_index = pd.read_excel('CTG.xls', sheet_name="Raw Data")
df_sheet_index = df_sheet_index.iloc[1:]
data = df_sheet_index.iloc[1:df_sheet_index.shape[0]-3]
data
```

```python
data["NSP"] = np.where(data["NSP"]<2,1,0)
data
```

```python
df_data = {"Model":["naive bayesian", "decision tree", "random forest"], "TP":[], "FP":[], "TN":[],"FN":[], "accuracy":[], "TPR": [], "TNR":[]}
```

```python
data = data[["LB", "ALTV", "Min", "Mean", "NSP"]]
X_train, X_test,y_train, y_test = train_test_split(data.iloc[:,:-1],data.iloc[:,-1] ,random_state=64, test_size=0.5, shuffle=True)
```

```python
model = MultinomialNB()
model.fit(X_train.to_numpy(), y_train.to_numpy())
y_pred = model.predict(X_test.to_numpy())
temp = X_test.copy()
temp["NSP"] = y_test
temp["predict"] = y_pred
temp
```

```python
accuracy_score(y_test.to_numpy(), y_pred)
```

```python
mat = confusion_matrix(y_test,y_pred)
sns.heatmap(mat.T,square=True, annot=True, fmt = 'd', cbar=True, xticklabels=["Normal", "Abnormal"], yticklabels=["Normal", "Abnormal"])
plt.xlabel('true label')
plt.ylabel('predicted label')
```

```python
tp = np.count_nonzero((y_pred==1) & (y_test==1))
fp = np.count_nonzero((y_pred==1) & (y_test==0))
tn = np.count_nonzero((y_pred==0) & (y_test==0))
fn = np.count_nonzero((y_pred==0) & (y_test==1))
print("TP -",tp,", FP -",fp,", TN -",tn,", FN - ",fn)
print("TPR -",tp/(tp+fn))
print("TNR -",tn/(tn+fp))
df_data["TP"].append(tp)
df_data["FP"].append(fp)
df_data["TN"].append(tn)
df_data["FN"].append(fn)
df_data["accuracy"].append(accuracy_score(y_test.to_numpy(), y_pred))
df_data["TPR"].append(tp/(tp+fn))
df_data["TNR"].append(tn/(tn+fp))
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X_train, y_train)
y_pred = tree_clf.predict(X_test)
temp = X_test.copy()
temp["NSP"] = y_test
temp["predict"] = y_pred
temp
```

```python
d_array = []
print("Random Forest Classifier - ")
m = [-1,[0,0]]
for d in range(1,6):
    d_array.append([[],[]])
    for n in range(1,11):
        rnd_clf = RandomForestClassifier(n_estimators=n, max_depth=d, warm_start=True,oob_score=True, criterion="entropy")
        rnd_clf.fit(X_train, y_train)
        y_pred = rnd_clf.predict(X_test)
        if accuracy_score(y_test.to_numpy(), y_pred)>m[0]:
            m[1][0] = n
            m[1][1] = d
            m[0] = accuracy_score(y_test.to_numpy(), y_pred)
        d_array[d-1][0].append(n)
        d_array[d-1][1].append(1- accuracy_score(y_test.to_numpy(), y_pred))
        print("N -", n, "Max depth d -", d, " Error Rate -", str(1- accuracy_score(y_test.to_numpy(), y_pred)))

for d in range(1,6):
    plt.plot(d_array[d-1][0], d_array[d-1][1], label = "d - "+str(d))
# print(d_array)
plt.xlim(0, 12)
plt.xlabel("N - number of (sub)trees ")
plt.ylabel("Error Rates")
plt.legend(loc="upper right")
plt.show()
print("Random Forest Max accuracy -", m[0], "for N -", m[1][0], "and d -", m[1][1])
```

```python
rnd_clf = RandomForestClassifier(n_estimators=m[1][0], max_depth=m[1][1], warm_start=True,oob_score=True, criterion="entropy")
rnd_clf.fit(X_train, y_train)
y_pred = rnd_clf.predict(X_test)
accuracy_score(y_test.to_numpy(), y_pred)
mat = confusion_matrix(y_test,y_pred)
sns.heatmap(mat.T,square=True, annot=True, fmt = 'd', cbar=True, xticklabels=["Normal", "Abnormal"], yticklabels=["Normal", "Abnormal"])
plt.xlabel('true label')
plt.ylabel('predicted label')
```
